#### Customize simple x-y graph

```r
#plot(carb$pH,carb$pCO2,col="red")
plot(H,carb$pCO2,col="red", xlab="H+ (total scale)", 
ylab="pCO2 (uatm)")
	
```

#### Simple contour plot: pCO2 as a function of the input pair pH-At

```r
# One wants to know pCO2 at various pHs and ATs:
pH <- seq(7.9, 8.2, length.out=10)
AT <- seq(2200e-6, 2600e-6, length.out=10)
dat <- expand.grid(pH, AT)
dat
carb <- carb(flag=8, var1=dat$Var1, var2=dat$Var2, S=35, T=25, P=0, Pt=0, Sit=0, k1k2="l", kf="pf", pHscale="T")
carb

# Do a simple contour plot
pCO2 <- carb$pCO2
dim(pCO2) <- c(length(pH), length(AT)) # 
contour(pH, AT, pCO2)
```

#### More elaborated contour plot

```r
contour(pH, AT*1e6, pCO2, 
	xlab="pH (total scale)",
	ylab="Total alkalinity (umol/kg)",
	levels=seq(200, 700, by=100),  
	labcex=1.5,
	method="edge",
	col="red",
	lwd=2,
	lty="solid",
	)
```

#### Convert pH to a different scale

```r
# SWS to total scale - a single value
pHconv(flag=1, pH=8.10, S=35, T=25, P=0)

# Multiple values in one go
pHsws <- seq(7.7, 8.3, by=0.01)	
pHsws
pHtotal <- pHconv(flag=1, pH=pHsws, S=35, T=25, P=0)
pHtotal
```

## 3. Effect of pure water addition on ocean pH

#### Reference state (roughly Arctic conditions): $pCO_2=400$ $\mu$atm and  $A_T=2325$ $\mu$mol/kg

```r
# pCO2 and AT are known:
ref <- carb(flag=24, var1=400, var2=2325e-6, S=34, T=2, P=0, Pt=0, Sit=0, k1k2="w14", kf="pf", pHscale="T")
refDIC = ref$CO2 + ref$HCO3 + ref$CO3
ref
```

#### Test to assure that consistent results are achieved with a different pair (ALK, DIC): yes

```r
# pCO2 and AT are known:
carb(flag=15, var1=2325e-6, var2=refDIC, S=34, T=2, P=0, Pt=0, Sit=0, k1k2="w14", kf="pf", pHscale="T")
```

#### Effect for a dilution of $A_T$, $C_T$, and $S$ to half their original values: pH increases

```r
# pCO2 and AT are known:
a=carb(flag=15, var1=(2325e-6)/2, var2=refDIC/2, S=34/2, T=2, P=0, Pt=0, Sit=0, k1k2="w14", kf="pf", pHscale="T")
```

pH change from dilution + equlibration with atmospheric CO2

```r
a$pH - ref$pH
```

#### Effect from then equilibrating the diluted sample with atmospheric CO$_2$ (at 400 $\mu$atm): pH declines (dominant effect)

```r
# pCO2 and AT are known:
b <- carb(flag=24, var1=400, var2=2325e-6/2, S=34/2, T=2, P=0, Pt=0, Sit=0, k1k2="w14", kf="pf", pHscale="T")
b
```

```r
b$pH - ref$pH
```
